```python
def label_encoder(my_dataframe, columns_to_encode: List[str]) -> pd.DataFrame:
    for column_to_encode in (columns_to_encode):
        try:
            if column_to_encode in my_dataframe.columns:
                try:
                    if my_dataframe[column_to_encode].dtypes==object:
                        tmp = my_dataframe[column_to_encode].astype('category')
                        dictionary = dict(enumerate(tmp.cat.categories))
                        reverse_dictionary = dict([(value, key) for key, value in dictionary.items()])
                        updated_frame = my_dataframe[column_to_encode].map(reverse_dictionary)
                        my_dataframe[column_to_encode]= updated_frame
                        print("{} column has been successfully encoded.".format(column_to_encode))
                    else:
                        raise Exception;
                except Exception:
                    print("{} is not a valid column to be encoded.".format(column_to_encode))
            else:
                raise ValueError
        except ValueError:
            print("There is no column named {} in the dataset.".format(column_to_encode))
print(my_dataframe)
return my_dataframe
```
